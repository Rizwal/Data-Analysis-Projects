# DISTRIBUTION PLOTS(distplots) FOR NUMERIC VALUES

```python
sns.displot(x='Age',data=df,kde=True)
```

```python
sns.distplot(df['Age'],hist=True,kde=False)
```

```python
sns.distplot(df['Age'],hist=False,kde=False,rug=True)
```

```python
sns.distplot(df['Age'],hist=True,kde=False,bins=5)
```

```python
sns.distplot(df['RestingBP'],color='red')
```

```python
sns.distplot(df['Cholesterol'],color='red')
```

```python
sns.distplot(df['FastingBS'],color='red')
```
